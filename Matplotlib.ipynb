### **Exercise**
Given score points of Men and Women in a certain game played 5 times with the corresponding errors in the observations.<br>
<br>
Plot a bar graph depicting scores for all 5 games along  with errors in observation. Take colors of your choice.

```python
n_grps = np.arange(5)
bar_width = 0.4
men_sc = [20, 30, 10, 50, 90]
err_men_sc = [2, 3, 4, 5, 4]

women = [10, 123, 19, 60, 40]
err_women_sc = [1, 6, 2, 8, 7]
```

### **Pie chart**

```python
df_pie = df[['Type_1','Attack','Defense','Speed','HP']].copy()
print(df_pie.head())
```

```python
frequent_grp = np.unique(df_pie.iloc[:,0],return_counts=True)
frequent_grp = np.array(frequent_grp).T

frequent_grp = frequent_grp[frequent_grp[:,1].argsort()[::-1]][0:4,:]
print(frequent_grp)
```

```python
df_pie= df_pie.loc[df_pie.loc[:,'Type_1'].str.contains(r'(Water|Normal|Grass|Bug)')]
print(df_pie)
```

```python
#names of the group
type_1_names=frequent_grp[:,0]
print(type_1_names)

#mean of samples for each feature corresponding to all 4 groups
df_grp = df_pie.groupby('Type_1').mean()
print(df_grp)
```

```python
names= df_grp.columns
colors=['gold','lightcoral','yellowgreen','lightskyblue']
explode = (0,0,0,0.1)
fig, [[ax1,ax2],[ax3,ax4]]=plt.subplots(nrows=2,ncols=2)
ax=[ax1, ax2, ax3, ax4]
for i in range(0,4):
    percent = df_grp.iloc[i,:]
    ax[i].pie(percent,explode=explode,labels=names,colors=colors,autopct='%.2f%%',shadow=True,startangle=90)
    ax[i].set_aspect('equal')
    ax[i].set_title(type_1_names[i])

plt.suptitle('Comparing major features of 4 most frequent Pokemon Group',fontsize=14,fontweight='bold')
```

### **Animated graphs**

### **Line Chart**

```python
df = pd.read_csv('/content/drive/MyDrive/pokemon_alopez247.csv')
```

```python
bulbasaur = list(df.iloc[0,4:11])
charmander = list(df.iloc[3,4:11])
squirtle = list(df.iloc[6,4:11])
pokets = [bulbasaur, charmander, squirtle]

#converting height from meters to inches
bul_hw = [df.iloc[0,19]*3.281,df.iloc[0,20]]
char_hw = [df.iloc[3,19]*3.281,df.iloc[3,20]]
squir_hw = [df.iloc[6,19]*3.281,df.iloc[6,20]]

hw = [bul_hw, char_hw, squir_hw]
```

```python
fig = plt.figure(0)

#line plot
ax1 = plt.subplot2grid((4,3),(0,0),colspan=3)

ax1.plot(bulbasaur,'g-',bulbasaur,'go')
ax1.plot(charmander,'r-',charmander,'ro')
ax1.plot(squirtle,'b-',squirtle,'bo')
ax1.set_xlim(-0.3,6.3)
```

```python
fig = plt.figure(0)

#line plot
ax1 = plt.subplot2grid((4,3),(0,0),colspan=3)

ax1.plot(bulbasaur,'g-',bulbasaur,'go')
ax1.plot(charmander,'r-',charmander,'ro')
ax1.plot(squirtle,'b-',squirtle,'bo')
ax1.set_xlim(-0.3,6.3)

text_x_coord = [0.045, .19, 0.34, 0.49, 0.64, 0.795, 0.945]
text_y_coord = [0.7, 0.57, 0.65, 0.75, 0.75, 0.75, 0.72]
rot = [0, 45, 45, 45, 45, 45, 45]
txt = [r'$T_{OTAL}$',r'$HP$',r'$A_{TTACK}$',r'$D_{EFENSE}$',r'$S_{P}\_A_{TK}$',r'$S_{P}\_D_{EF}$',r'$S_{PEED}$']

for i in range(0,7):
    ax1.text(text_x_coord[i], text_y_coord[i], txt[i], transform=ax1.transAxes,va='top',rotation=rot[i],bbox=dict(boxstyle='round',facecolor='wheat',alpha=0.78))
```

### **Box Plot**

```python
ax2 = plt.subplot2grid((4,3),(1,0),colspan=1)
ax3 = plt.subplot2grid((4,3),(2,0),colspan=1)
ax4 = plt.subplot2grid((4,3),(3,0),colspan=1)
ax = [ax2, ax3, ax4]

colors = ['g', 'r', 'b']

for i in range(0,3):
    bp=ax[i].boxplot(pokets[i][1:],patch_artist=True)
    # adding colors to edges
    for element in ['boxes','whiskers','fliers','means','medians','caps']:
        plt.setp(bp[element],color='k')
    #adding color inside the box
    for patch in bp['boxes']:
        patch.set(facecolor=colors[i])
    ax[i].set_ylim(35,70)

ax3.annotate('Median',xy=(1.09,51),xytext=(1.2,60),arrowprops=dict(facecolor='wheat',shrink=0.001),)
```

### **Plot Texts - Exercise**

```python
x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2*np.pi*x)*np.exp(-x)
```

```python
#plot
plt.figure(figsize=(8,5))

#line + markers
plt.plot(x,y,'k-',linewidth=1.5)
plt.scatter(x,y,color='c',s=30)

#titles
plt.suptitle("Main heading", fontsize=14, fontweight='bold')
plt.title("Figure Title",fontsize =12)

#axis labels
plt.xlabel("Abscissa",color='green')
plt.ylabel('Ordinate',color='green')

#grid
plt.grid(True)

#annotation: theta: 60 degrees
plt.text(0.95, 0.35, r'$\theta=60^\circ$',fontsize=10)

#annotation: 2nd crest with arrow
crest_x = 2
crest_y = np.cos(2*np.pi*crest_x)*np.exp(-crest_x)

plt.annotate("2nd crest", xy=(crest_x, crest_y),xytext=(2.5,0.3),arrowprops=dict(facecolor='green',shrink=0.05,width=2),fontsize=10)

plt.show()
```
