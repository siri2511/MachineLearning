### **Exercise**

1. write a python code to separate this data into distinct columns
2. provide header to the dataframe with given values
3. list names of the column along with count of missing number of values. missing values coded with '?'
4. find the column names ending with either of the following 'N', 'H', 'S'. if the number of recorded data in these columns are less than 100, drop them
5. drop all the rows which have more than or equal to 2 missing values. What is the shape of the final dataframe
6. fill all the remaining missing values with the mode of each column

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/bridges/bridges.data.version1'
columns=['IDENTIF','RIVER','LOCATION','ERECTED','PURPOSE','LENGTH','LANES','CLEAR-G','T-OR-D','MATERIAL','SPAN','REL-L','TYPE']
df = pd.read_csv(url,names=columns)
print(df.head())
print(df.shape)
```

```python
df.replace('?',np.nan, inplace=True)
missing_counts = df.isna().sum()
print(missing_counts)
```

```python
cols_to_check = df.columns[df.columns.str.endswith(('N', 'H', 'S'))]

for col in cols_to_check:
    if df[col].count()<100:
        df.drop(columns=col,inplace=True)

print(df.columns)
```

```python
df = df.dropna(thresh=df.shape[1]-1)
print(df.shape)
```

```python
for col in df.columns:
    df[col].fillna(df[col].mode()[0],inplace=True)

print(df.head())
```

### **Pivot Tables**

```python
df = pd.DataFrame([['IND','Gold','Game1','9.9'],['IND','Bronze','Game2','8'],['USA','Silver','Game1','9.5'],['USA','Gold','Game2','8.6']],columns=['Country','Medal','Game','Score'],index=['Year1','Year2','Year1','Year2'])
print(df)
```

Pivot tables: breaking down a large dataset into fewer features for quick visualisation

```python
# listing all the features
print(df.pivot(index='Country',columns='Medal'))
```

```python
# listing only score feature
print(df.pivot(index='Country',columns='Medal',values='Score'))
```

pivot_table(): requires argument as aggfun through which we can clear the ambiguity

```python
df = pd.DataFrame([['IND','Gold','Game1','9.9'],['IND','Bronze','Game2','8'],['USA','Silver','Game1','9.5'],['USA','Gold','Game2','8.6']],columns=['Country','Medal','Game','Score'],index=['Year1','Year2','Year1','Year2'])
print(df)
```

```python
print(df.pivot_table(index='Country',columns='Game',values='Score',aggfunc=np.max))
```

### **Grouping - Exercise**
1. construct a dataframe new_df where the given dataset is grouped based on each system (s1 and s2) and speed per day (d1 and d2) with the median speed each day per system. Also, provide a secondary name 'Median' for the speed attribute.
2. Sort the dataframe new_df in the ascening order of the median speed.

```python
sys=['s1','s1','s1','s1','s2','s2','s2','s2']
net_day = ['d1','d1','d2','d2','d1','d1','d2','d2']
spd = [1.3,11.4,5.6,12.3,6.2,1.1,20.0,8.8]
df = pd.DataFrame({'set_name':sys,'spd_per_day':net_day,'speed':spd})
print(df)
```

```python
new_df = df.groupby(['set_name','spd_per_day'],sort=True).median().rename(columns={'speed':'Median'})
print(new_df)
```
