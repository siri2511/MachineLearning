### **Operations on axis - Exercise**
Given a matrix mat of size 3x3. Find the minimum numbers from each row, say N1, N2, N3. Result in a mtrix by adding:<br>
1. N1 to the upper half elements of mat
2. N2 to the main diagonal elements of the mat
3. N3 to the lower half elements of mat

```python
mat = np.array([[10,5,9],[2,20,6],[8,3,30]]).reshape(3,3)
N1_ind = np.argmax(mat,axis=0)
```

```python
print(N1_ind)
```

```python
add_mat = np.zeros((3,3))
add_mat[np.triu_indices_from(add_mat, k=1)]=mat[0][N1_ind[0]]

np.fill_diagonal(add_mat,mat[1][N1_ind[1]])

add_mat[np.tril_indices_from(add_mat, k=-1)]=mat[2][N1_ind[2]]


mat = mat+add_mat
print(mat)
```

### **Matrix indexing & Slicing - Exercise**
1. perform black and white thresholding with decison boundary set at 0.15. store the new image in a variable new_img and show it.
2. use the numpy array manipulating function fliplr() to get the mirror image
3. compress the flipped image along both the axes by 50% in sequence.

```python
from skimage.data import shepp_logan_phantom

phantom = shepp_logan_phantom()

plt.imshow(phantom,cmap='gray')
plt.title('Shepp-Logan Phantom')
plt.axis('off')
plt.show()
```

```python
new_img = np.where(phantom<0.15,0,1)
plt.imshow(new_img,cmap='gray')
plt.title('New Shepp-Logan Phantom')
plt.axis('off')
plt.show()
```

```python
mirror_img = np.fliplr(new_img)
plt.imshow(mirror_img,cmap='gray')
plt.title('Mirrored New Shepp-Logan Phantom')
plt.axis('off')
plt.show()
```

```python
rows, cols = mirror_img.shape

mask_x = np.arange(cols)%2==0
mask_y = np.arange(rows)%2==0

compressed_x = np.compress(mask_x,mirror_img, axis=1)

compressed_xy = np.compress(mask_y, compressed_x, axis=0)

plt.imshow(compressed_xy,cmap='gray')
plt.title('Compressed Mirrored New Shepp-Logan Phantom')
plt.axis('off')
plt.show()
```

### **Shape Manipulation**

```python
from skimage.data import immunohistochemistry

ihc_img = immunohistochemistry()

plt.imshow(ihc_img)
plt.title('IHC Image')
plt.axis('off')
plt.show()
```

```python
#splitting a matrix

h_slice = ihc_img.copy()
v_slice = ihc_img.copy()

h_slice = np.hsplit(h_slice,2)
v_slice = np.vsplit(v_slice,4)

#horizontal split
plt.figure()
plt.imshow(h_slice[1])

```

```python
#vertical split
plt.figure()
plt.imshow(v_slice[1])
```

```python
#merging matrices
h_stack=np.hstack((h_slice[1],h_slice[0]))
v_stack = np.vstack((v_slice[1],v_slice[0]))

#horizontal merge
plt.figure()
plt.imshow(h_stack)
```

```python
#vertical merge
plt.figure()
plt.imshow(v_stack)
```

### **Linear Algebra**
<br>
**linalg** : module for linear algebra

```python
mat1 = np.arange(4).reshape(2,2)
mat2 = (np.arange(4)*2).reshape(2,2)
mat3 = (np.arange(4)*3).reshape(2,2)

#performing dot product
print(np.linalg.multi_dot([mat1, mat2, mat3]))
```

Consider a system of linear equations:<br>
<br>
3x + y = 9<br>
x + 2y = 8
<br>
<br>
if the system has unique solutions, find inverse and rank of the matrix so formed.

```python
a = np.array([[3,1],[1,2]])
b = np.array([9,8])

#check if unique solution exists
print(np.linalg.det(a))

#det = 5 -> unique solution
print(np.linalg.solve(a,b))

#calculating inverse (det!=0 -> matrix invertible)
print(np.linalg.inv(a))

#rank calculation
print(np.linalg.matrix_rank(a))
```
